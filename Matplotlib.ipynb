#Matplotlib


* Matplotlib is a powerful and versatile open-source plotting library for Python, designed to help users visualize data in a variety of formats.
* Developed by John D. Hunter in 2003, it enables users to graphically represent data, facilitating easier analysis and understanding.
* If you want to convert your boring data into interactive plots and graphs, Matplotlib is the tool for you.

```python
#Importing Matplotlib
import matplotlib.pyplot as plt
```

```python
# Line Plot
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
plt.title('Line Plot Example')
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
#Scatter Plot
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.scatter(x, y)
plt.title('Scatter Plot Example')
plt.xlabel('x values')
plt.ylabel('y values')
plt.show()
```

```python
#Bar Plot
import matplotlib.pyplot as plt
import numpy as np
categories = ['A', 'B', 'C', 'D']
values = [5, 10, 15, 20]
plt.bar(categories, values)
plt.title('Bar Plot Example')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.show()
```

```python
#Histogram
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(1000)
plt.hist(data, bins=30, edgecolor='black')
plt.title('Histogram Example')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
