#Q) What are operators in Python

* Operators in Python are special symbols that perform operations on variables and values.
* They can be classified into several categories based on their functionality.

* 1.Arithmetic Operators:
These operators are used to perform mathematical operations.

* 2.Comparison Operators:
These operators compare two values and return a boolean result (True or False).

* 3.Logical Operators:
These operators are used to combine conditional statements

* 4.Assignment Operators:
These operators are used to assign values to variables.

* 5.Identity Operators:
These operators compare the memory locations of two objects.

* 6.Membership Operators:
These operators test for membership in a sequence .

#Q) What are Logical OPerators in Python?

*  Logical Operators:
These operators are used to combine conditional statements.

* 1.AND (and): Returns True if both operands are true.

* 2.OR (or): Returns True if at least one of the operands is true.

* 3.NOT (not): Reverses the boolean value of the operand.
