* It provides beautiful default styles and color palettes to make statistical plots more attractive.
* It is built on top matplotlib library and is also closely integrated with the data structures from pandas.
* Seaborn aims to make visualization the central part of exploring and understanding data.

```python
#Importing Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
```

#Seaborn Dataset
* Seaborn comes with several built-in datasets that are useful for practicing and creating visualizations.
* You can load these datasets using the seaborn.load_dataset() function.

```python
#seaborn datasets
sns.get_dataset_names()
```

```python
#Line Plot
tips = sns.load_dataset("tips")
sns.lineplot(x="day", y="total_bill", data=tips)
plt.title('Total Bill vs Day')
plt.show()
```

```python
#Scatter Plot
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title('Total Bill vs Tip')
plt.show()
```

```python
#Bar Plot
sns.barplot(x="day", y="total_bill", data=tips)
plt.title('Average Total Bill by Day')
plt.show()
```

```python
#Histogram
sns.histplot(tips['total_bill'], kde=True)
plt.title('Total Bill Distribution')
plt.show()
```

```python
#Box Plot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title('Total Bill Distribution by Day')
plt.show()

```

```python
#Violin plot
sns.violinplot(x="day", y="total_bill", data=tips)
plt.title('Total Bill Distribution by Day')
plt.show()
```

```python
#Regression plot
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title('Regression: Total Bill vs Tip')
plt.show()
```
